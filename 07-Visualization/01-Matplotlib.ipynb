# Introduction to Matplotlib

Matplotlib is a powerful and versatile Python library widely used for creating static, animated, and interactive visualizations in Python. It serves as the foundational plotting library upon which many other visualization libraries, such as Seaborn, are built. Understanding Matplotlib is essential for anyone looking to perform data analysis and visualization in Python.

## Overview

Matplotlib provides a comprehensive suite of tools for creating a wide variety of plots and charts. Whether you're looking to generate simple line plots or complex 3D visualizations, Matplotlib offers the flexibility and control needed to produce high-quality graphics.

### Key Features of Matplotlib

- **Versatility:** Supports a wide range of plot types including line plots, scatter plots, bar charts, histograms, pie charts, and more.
- **Customization:** Highly customizable, allowing users to tweak every aspect of a plot, from colors and fonts to axes and legends.
- **Integration:** Seamlessly integrates with other Python libraries such as NumPy, Pandas, and SciPy, making it a staple in the data science ecosystem.
- **Extensibility:** Users can extend Matplotlib with additional libraries and tools to create more complex visualizations.

## Importance in Data Visualization

Data visualization is a critical component of data analysis, as it allows for the effective communication of insights and findings. Matplotlib plays a pivotal role in this process for several reasons:

### 1. **Clarity and Insight**

Visual representations of data can reveal patterns, trends, and outliers that may not be immediately apparent from raw data alone or even from the summary statistics. Matplotlib enables users to create clear and informative visualizations that enhance understanding.

### 2. **Customization and Control**

With Matplotlib, users have granular control over every element of a plot. This level of customization ensures that visualizations can be tailored to meet specific needs and preferences, making the data more accessible and interpretable.

### 3. **Foundation for Advanced Libraries**

Many advanced visualization libraries, such as Seaborn and Plotly, are built on top of Matplotlib. A solid understanding of Matplotlib provides a strong foundation for leveraging these higher-level libraries to create more sophisticated visualizations with ease.

### 4. **Community and Support**

Matplotlib boasts a large and active community, offering extensive documentation, tutorials, and support. This makes it easier for beginners to learn and for experienced users to troubleshoot and discover new techniques.

## Getting Started with Matplotlib

Let's dive into a simple example to illustrate how Matplotlib works. We'll create a basic line plot to visualize a mathematical function.

### Example: Plotting a Sine Wave

````python
from IPython.display import Pretty as disp
hint = 'https://raw.githubusercontent.com/soltaniehha/Business-Analytics/master/docs/hints/'  # path to hints on GitHub

# Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2Ï€
y = np.sin(x)                        # Sine of each x point

# Create the plot
plt.figure(figsize=(8, 4))           # Set the figure size
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Customize the plot
plt.title('Sine Wave')               # Add a title
plt.xlabel('Angle [radians]')        # Label for the x-axis
plt.ylabel('sin(x)')                 # Label for the y-axis
plt.legend()                         # Add a legend
plt.grid(True)                       # Show grid lines
````

At the core of Matplotlib's functionality is the ability to create simple line plots using the `plt.plot()` function. You can pass in data points (e.g., arrays or lists) for the x and y axes, and Matplotlib will automatically generate a line connecting those points. To display the plot, use `plt.show()`, which renders the figure. In more recent versions of Jupyter notebook (or Colab) this is optional as the notebook will automatically display the plot.

## Figure and Axes

In Matplotlib, the **Figure** and **Axes** objects are fundamental components that form the backbone of any plot. Understanding these objects allows you to create more complex and customized visualizations.

### Understanding the Figure and Axes Objects

- **Figure**: This is the overall container for your plot. It can contain multiple Axes (plots), along with other elements like titles, legends, and colorbars.

- **Axes**: These are the individual plots within a Figure. Each Axes has its own set of data, labels, and styles.

### Creating Subplots with `plt.subplots()`

The `plt.subplots()` function is a convenient way to create a Figure along with one or more Axes. This is especially useful when you want to display multiple plots side by side or in a grid layout.

#### Example: Creating a Single Axes using object-oriented interface

The object-oriented interface is available for more complicated situations, and for when you want more control over your figure.
Rather than depending on some notion of an "active" figure or axes, **in the object-oriented interface the plotting functions are *methods* of explicit ``Figure`` and ``Axes`` objects**.

The main difference between `plt.plot()` and `ax.plot()` is that `plt.plot()` uses a simpler, functional interface that automatically manages the figure and axes, making it ideal for quick plots. In contrast, `ax.plot()` is part of an object-oriented approach where you explicitly create and manage the axes, offering more control and flexibility, especially for complex visualizations or when working with multiple subplots.

Below we re-create the previous plot using this style of plotting:

````python
# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 4))  # Set the figure size

# Plot the data on the axes
ax.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Customize the plot
ax.set_title('Sine Wave')                   # Add a title
ax.set_xlabel('Angle [radians]')            # Label for the x-axis
ax.set_ylabel('sin(x)')                     # Label for the y-axis
ax.legend()                                 # Add a legend
ax.grid(True)                               # Show grid lines
````

**Explanation:**

1. **Creating Figure and Axes**:
    ```python
    fig, ax = plt.subplots(figsize=(8, 4))
    ```
    - `fig`: Represents the Figure object.
    - `ax`: Represents the Axes object where the data will be plotted.

2. **Plotting Data**:
    ```python
    ax.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)
    ```
    - Plots the sine wave on the Axes.

3. **Customizing the Plot**:
    - **Title and Labels**:
        ```python
        ax.set_title('Sine Wave')
        ax.set_xlabel('Angle [radians]')
        ax.set_ylabel('sin(x)')
        ```
    - **Legend and Grid**:
        ```python
        ax.legend()
        ax.grid(True)
        ```

### Creating Multiple Subplots

Using `plt.subplots()`, you can create multiple Axes within a single Figure. This is useful for comparing different datasets or different views of the same data.

#### Example: Side-by-Side Plots


````python
# Generate additional data
y2 = np.cos(x)  # Cosine of each x point

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns

# First subplot: Sine Wave
axes[0].plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('Angle [radians]')
axes[0].set_ylabel('sin(x)')
axes[0].legend()
axes[0].grid(True)

# Second subplot: Cosine Wave
axes[1].plot(x, y2, label='Cosine Wave', color='red', linestyle='--', linewidth=2)
axes[1].set_title('Cosine Wave')
axes[1].set_xlabel('Angle [radians]')
axes[1].set_ylabel('cos(x)')
axes[1].legend()
axes[1].grid(True)

# Adjust layout and display the plots - Automatically adjusts subplot parameters to give specified padding.
plt.tight_layout()
````

## Plot Customization

As illustrated in the examples above, Matplotlib allows you to customize your plots by adding titles, axis labels, and setting limits for the x- and y-axes. These simple customizations make your plots more informative and visually appealing.

### Adding Titles and Labels

You can add a title to your plot using `plt.title()` and label the x- and y-axes using `plt.xlabel()` and `plt.ylabel()` respectively. You can also use the `set_title()`, `set_xlabel()`, and `set_ylabel()` on the "axes" objects, when using the object-oriented interface.

### Setting Axis Limits

You can control the limits of the x- and y-axes using `plt.xlim()` and `plt.ylim()`. This is useful when you want to zoom in on a specific region of the plot.

#### Example:


````python
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Customize the plot with axis limits
plt.title('Zoomed Sine Wave')         # Add a title
plt.xlabel('Angle [radians]')         # Label for the x-axis
plt.ylabel('sin(x)')                  # Label for the y-axis
plt.xlim(0, np.pi)                    # Set x-axis limit from 0 to Ï€
plt.ylim(-0.5, 1)                     # Set y-axis limit from -0.5 to 1
plt.legend()                          # Add a legend
plt.grid(True)                        # Show grid lines
````

## Styling Plots

Matplotlib offers many options to style your plots. You can modify line styles, colors, and markers to make your plots more visually distinct. Additionally, Matplotlib provides built-in plot styles to easily change the overall appearance of your visualizations.

### Line Styles, Colors, and Markers
You can change the appearance of the plot's lines by modifying line style, color, and adding markers.

#### Common Customizations:
- **Line Style**: Use `linestyle='-'` for solid, `'--'` for dashed, `':'` for dotted.
- **Color**: Set the color with `color='blue'` or using color codes (e.g., `'r'` for red).
- **Markers**: Markers represent data points, e.g., `marker='o'` for circles, `marker='s'` for squares.

#### Example:


````python
# Create the plot
plt.figure(figsize=(8, 4))           # Set the figure size
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', marker='o', linewidth=1)  # Solid line with circles

# Customize the plot
plt.title('Styled Sine and Cosine Waves')  # Add a title
plt.xlabel('Angle [radians]')              # Label for the x-axis
plt.ylabel('Value')                        # Label for the y-axis
plt.legend()                               # Add a legend
plt.grid(True)                             # Show grid lines
````

### Using Plot Styles

Matplotlib offers several built-in plot styles that allow you to quickly change the overall look of your plots. You can use `plt.style.use()` to apply these styles globally across all plots.

Some common styles include `'ggplot'`, `'seaborn'`, and `'fivethirtyeight'`. See `plt.style.available` for available style options in your environment.

#### Example:


````python
# Apply a style
plt.style.use('ggplot')  # Apply the 'ggplot' style

# Create the plot
plt.figure(figsize=(8, 4))           # Set the figure size
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', marker='o', linewidth=2)  # Solid line with circles

# Customize the plot
plt.title('Styled Sine and Cosine Waves with ggplot')  # Add a title
plt.xlabel('Angle [radians]')                          # Label for the x-axis
plt.ylabel('Value')                                    # Label for the y-axis
plt.legend();                                          # Add a legend
````

In this example, we applied the `'ggplot'` style to the plot using `plt.style.use('ggplot')`, which modifies the colors and layout to resemble the popular ggplot style from R. You can experiment with different styles to suit your visual preference.

**Note:** The ";" at the end of the plot command will stop the unnecessary object to be printed.

## Basic Plot Types

Matplotlib provides several basic plot types that are useful for different kinds of data visualizations. In this section, we will look at three common types: scatter plots, bar charts, and histograms.

### Scatter Plots (`plt.scatter()`)
Scatter plots are used to display values for typically two variables for a set of data. This plot helps visualize the relationship between the two variables.

#### Example:


````python
# Generate data for scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 100 * np.random.rand(50)      # Bubble sizes

# Create the scatter plot
plt.figure(figsize=(12, 6))           # Set the figure size
plt.scatter(x, y, s=sizes, c=y, alpha=0.5, cmap='viridis')

# Customize the plot
plt.title('Scatter Plot')             # Add a title
plt.xlabel('X Values')                # Label for the x-axis
plt.ylabel('Y Values')                # Label for the y-axis
plt.colorbar();                       # Show color scale
````

### Bar Charts (`plt.bar()`)
Bar charts are used to represent categorical data with rectangular bars. The height (or length) of the bars corresponds to the value of the data.

#### Example:


````python
# Data for bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Create the bar chart
plt.figure(figsize=(12, 6))           # Set the figure size
plt.bar(categories, values, color='purple')

# Customize the plot
plt.title('Bar Chart')               # Add a title
plt.xlabel('Categories')             # Label for the x-axis
plt.ylabel('Values');                # Label for the y-axis
````

### Histograms (`plt.hist()`)
Histograms are used to represent the distribution of a dataset. The data is divided into bins, and the frequency of data within each bin is calculated and then visualized by the height of the bars.

#### Example:


````python
# Data for histogram
data = np.random.randn(1000)         # Generate 1000 random numbers

# Create the histogram
plt.figure(figsize=(12, 6))          # Set the figure size
plt.hist(data, bins=30, color='blue', alpha=0.7)

# Customize the plot
plt.title('Histogram')               # Add a title
plt.xlabel('Value')                  # Label for the x-axis
plt.ylabel('Frequency');             # Label for the y-axis
````

## Saving Figures

Matplotlib allows you to save your plots to files for later use. This is especially useful for sharing visualizations, including them in reports, or for publication purposes. The `plt.savefig()` function enables you to export your figures in various formats such as PNG, PDF, SVG, and more.

### Exporting Plots to Files (`plt.savefig()`)
You can use `plt.savefig()` to save your plots. This function allows you to specify the file format by using the appropriate file extension (e.g., `.png`, `.pdf`, `.svg`). You can also control the resolution, background color, and other settings.

#### Example:


````python
# Data for histogram
data = np.random.randn(1000)         # Generate 1000 random numbers

# Create the histogram
plt.figure(figsize=(12, 6))          # Set the figure size
plt.hist(data, bins=30, color='blue', alpha=0.7)

# Customize the plot
plt.title('Histogram')               # Add a title
plt.xlabel('Value')                  # Label for the x-axis
plt.ylabel('Frequency')              # Label for the y-axis

# Save the figure as a PNG file
plt.savefig('my_figure.png', dpi=300, bbox_inches='tight')  # Save the figure with high resolution (300 DPI)
````

## Annotations and Text

Annotations and text can help highlight important features or points in your plots. Matplotlib provides the `plt.text()` function for adding text at specific locations and the `plt.annotate()` function for more advanced annotations.

### Adding Text Annotations (`plt.text()`)
You can use `plt.text()` to place custom text at specific coordinates on the plot. This is useful for labeling specific data points or adding notes to your visualizations.

#### Example: Adding Text with `plt.text()`


````python
# Generate data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(12, 6))           # Set the figure size
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Add a text annotation
plt.text(np.pi, 0, 'Center of sine wave', fontsize=12, ha='center')

# Customize the plot
plt.title('Sine Wave with Text Annotation')  # Add a title
plt.xlabel('Angle [radians]')               # Label for the x-axis
plt.ylabel('sin(x)')                        # Label for the y-axis
plt.legend()                                # Add a legend
plt.grid(True)                              # Show grid lines
````

### Highlighting Specific Data Points
You can highlight important data points by using `plt.annotate()` or adding markers. `plt.annotate()` allows you to attach an annotation to a specific data point with an arrow pointing to it.

#### Example: Highlighting a Point with `plt.annotate()`


````python
# Create the plot
plt.figure(figsize=(12, 6))           # Set the figure size
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Highlight the maximum point on the sine wave
max_x = np.pi / 2
max_y = np.sin(max_x)
plt.annotate('Maximum', xy=(max_x, max_y), xytext=(max_x - 0.75, max_y - 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->', edgecolor='red', clip_on=False),
             fontsize=12)

# Customize the plot
plt.title('Sine Wave with Highlighted Maximum')  # Add a title
plt.xlabel('Angle [radians]')                    # Label for the x-axis
plt.ylabel('sin(x)')                             # Label for the y-axis
plt.legend()                                     # Add a legend
plt.grid(True)                                   # Show grid lines
````

````python
import datetime
import yfinance as yf

tickerSymbol = 'GOOGL'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
stock_df = tickerData.history(period = '1d',
                              start  = (datetime.date.today() - datetime.timedelta(days=3650)),
                              end    = datetime.date.today())

stock_df
````

Let's make a plot using the dataframe .plot() method:

````python
stock_close.plot();
````

Now, let's use subplots with the functional interface:

````python
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(stock_close)

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(stock_vol);
````

It is important to note that this interface is *stateful*: it keeps track of the "current" figure and axes, which are where all ``plt`` commands are applied.

While this stateful interface is fast and convenient for simple plots, it is easy to run into problems. For example, once the second panel is created, how can we go back and add something to the first?

This is possible within the MATLAB-style interface, but a bit clunky.
Fortunately, there is a better way.

#### Object-oriented interface

Re-creating the previous plot using this style of plotting:

````python
# First create a grid of plots
fig, ax = plt.subplots(2)  # ax will be an array of two Axes objects

# Call plot() method on the appropriate object
ax[0].plot(stock_close)
ax[1].plot(stock_vol);
````

For more simple plots, the choice of which style to use is largely a matter of preference, but the object-oriented approach can become a necessity as plots become more complicated.
We will switch between the MATLAB-style and object-oriented interfaces, depending on what is most convenient.
In most cases, the difference is as small as switching ``plt.plot()`` to ``ax.plot()``, but there are a few gotchas that we will highlight below:

## Aside: Matplotlib Gotchas

While most ``plt`` functions translate directly to ``ax`` methods (such as ``plt.plot()`` â†’ ``ax.plot()``, ``plt.legend()`` â†’ ``ax.legend()``, etc.), this is not the case for all commands.
In particular, functions to set limits, labels, and titles are slightly modified.
For transitioning between MATLAB-style functions and object-oriented methods, make the following changes:

- ``plt.xlabel()``  â†’ ``ax.set_xlabel()``
- ``plt.ylabel()`` â†’ ``ax.set_ylabel()``
- ``plt.xlim()``  â†’ ``ax.set_xlim()``
- ``plt.ylim()`` â†’ ``ax.set_ylim()``
- ``plt.title()`` â†’ ``ax.set_title()``

In the object-oriented interface to plotting, rather than calling these functions individually, it is often more convenient to use the ``ax.set()`` method to set all these properties at once:

````python
ax = plt.axes()
ax.plot(stock_close)
ax.set(xlim=((datetime.date.today() - datetime.timedelta(days=365*3)), datetime.date.today()),
       xlabel='Date', ylabel='Closing Price',
       title='{} Daily'.format(tickerSymbol))
plt.xticks(rotation=45);
````

### Best Practices for Visualizations

- **Choose the Right Plot Type:** Select chart types that best represent your data (e.g., bar charts for comparisons, line charts for trends).
- **Label Clearly:** Always include descriptive titles, axis labels, and legends to make your plots understandable.
- **Maintain Simplicity:** Avoid unnecessary decorations and clutter to highlight the key information.
- **Use Consistent Colors:** Apply a consistent color scheme and ensure colors are distinguishable, including for colorblind audiences.
- **Optimize Readability:** Use appropriate font sizes and styles, and ensure text is legible against the background.
- **Control Figure Size:** Adjust figure dimensions to enhance clarity and fit the presentation or publication format.
- **Leverage Gridlines Sparingly:** Use gridlines to aid interpretation without overwhelming the data.
- **Ensure Proper Scaling:** Use suitable scales (linear, logarithmic) to accurately represent data relationships.
- **Maintain Aspect Ratios:** Preserve aspect ratios when necessary to avoid distorting data representations.
- **Document Your Code:** Write clear comments and maintain organized code for reproducibility and ease of understanding.
- **Test for Accessibility:** Ensure visualizations are accessible to all users, including those with visual impairments.
- **Iterate and Refine:** Continuously review and improve your plots based on feedback and the clarity of data presentation.
